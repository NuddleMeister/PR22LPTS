kolk je posamezna bolnica nardila aka max (x plotovov)


```python
fig, plots = plt.subplots(1, 3)
fig.set_figwidth(20)

for name in dictHospitals["id"]:
    idString0 = "hospital." + name + ".bed.total"
    idString1 = "hospital." + name + ".icu.total"
    idString2 = "hospital." + name + ".care.total"
    if idString0 in hospitals:
        plots[0].plot(hospitals[idString0].fillna(0))
    if idString1 in hospitals:
        plots[1].plot(hospitals[idString1].fillna(0))
    if idString2 in hospitals:
        plots[2].plot(hospitals[idString2].fillna(0))

plots[0].grid()
plots[1].grid()
plots[2].grid()
```

```python
fig, plots = plt.subplots(1, 3)
fig.set_figwidth(20)

plots[0].plot(hospitals["hospital.bed.total"].fillna(0))
plots[1].plot(hospitals["hospital.icu.total"].fillna(0))
plots[2].plot(hospitals["hospital.care.total"].fillna(0))

plots[0].grid()
plots[1].grid()
plots[2].grid()
```
